```python
# Encapsulate the parsing logic
def getTextWaPo(url):
    page = urlopen(url).read().decode('utf8') # download the pag
    soup = BeautifulSoup(page,"lxml") # parse the page
    text = ' '.join(map(lambda p: p.text, soup.find_all('article'))) # artcile data
    return text.replace("?", "").encode('utf-8').decode('ascii','ignore')#text.encode('ascii', 'ignore') #text.encode('ascii', errors='replace').replace("?"," ")
```

```python
text = getTextWaPo(articleURL) # result is a string 
```

```python
# collect individual sentences into a list
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
text
```

```python
sents = sent_tokenize(text)
sents
```

```python
word_sent = word_tokenize(text.lower())
word_sent
```

```python
# Stopword -> nltk; punctuation -> python string module;
_stopwords = set(stopwords.words('english') + list(punctuation))
_stopwords
```

```python
word_sent=[word for word in word_sent if word not in _stopwords]
```

```python
word_sent
```

```python
# Construct a frequency distribution
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
```

```python
# Frequenct word
from heapq import nlargest # sort collections
```

```python
nlargest(10, freq, key=freq.get) # freq.get - value in the dctionary freq for guven key
```

```python
# Significance Score
# Keys -> Sentences; Values -> Significance Scores
# defaultdict - if you lookup a non-existing key, it's added to the dictionary
# ranking -> sentence dictionary -> sentences with their significance score

from collections import defaultdict
ranking = defaultdict(int)

# enumerate function -takes a list and returns a list of tuples, furst element is the
# index of the list element. 
# i -> index of the sentence, split sentence into words, lookup the frequency distribution
# table to identify frequency - importance of sentence based on word importances.
for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
            
ranking
```

```python
# Picking the top four significant sentences
sents_idx = nlargest(4, ranking, key=ranking.get)
sents_idx
```

```python
# sort the sentences and create the abstract
[sents[j] for j in sorted(sents_idx)]
```

```python
# Auto-summarization function
def summarize(text, n):
    sents = sent_tokenize(text)
    
    assert n <= len(sents) #check if the abstract len is less than or equal to document length
    word_sent = word_tokenize(text.lower())
    _stopwords = set(stopwords.words('english') + list(punctuation))
    
    word_sent=[word for word in word_sent if word not in _stopwords]
    freq = FreqDist(word_sent)
    
    
    ranking = defaultdict(int)
    
    for i,sent in enumerate(sents):
        for w in word_tokenize(sent.lower()):
            if w in freq:
                ranking[i] += freq[w]
             
        
    sents_idx = nlargest(n, ranking, key=ranking.get)
    return [sents[j] for j in sorted(sents_idx)]
```

```python
summarize(text,4)
```
